```python
import numpy as np
import pandas as pd
import itertools
```

```python
dist_matrix = [[100, 6, 15, 8, 13],[14, 100, 10, 11, 14],[5, 8, 100, 13, 11],[9, 8, 6, 100, 6],[5, 6, 15, 7, 100]]
```

```python
dist_df = pd.DataFrame(dist_matrix)
```

```python
def tsp(initial, dist_matrix):
    
    total_distance = 0
    for i in range(len(initial)):
        if i < len(initial)-1:
            row = initial[i]
            col = initial[i+1]
            total_distance += dist_matrix[row][col]
        else:
            row = initial[i]
            col = initial[0]
            total_distance += dist_matrix[row][col]

    return total_distance
```

```python
all_tours = list(itertools.permutations([0,1,2,3,4]))
```

```python
min_dist = 1000
iteration = 1
for tour in all_tours:
    new_dist = tsp(tour, dist_matrix)
    if new_dist < min_dist:
        min_tour = tour
        min_dist = new_dist
    print("Iter", iteration)
    print("Minimum Distance:", min_dist)
    print("Minimum Tour:", min_tour)
    print("Current Tour:", tour)
    print("-------------")
    iteration += 1
```
